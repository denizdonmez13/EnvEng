```python
#Deniz Donmez/1317210016
#Dilem Naz Erkan/1317210010
#Emine Dilara Kalpak/1317210024
#Irem Nur Kutuk/1317210011



import matplotlib.pyplot as plt

def calculate_cod(oxygen_initial, oxygen_final, factor, volume):
    cod_value = (oxygen_initial - oxygen_final) * factor * 8.36 * 1000 / volume
    return cod_value

def calculate_bod(oxygen_initial, oxygen_final, dilution_factor, incubation_days):
    bod_value = (oxygen_initial - oxygen_final) * dilution_factor * 8.36 * 1000 / incubation_days
    return bod_value

def calculate_nh4_n(nh4_concentration, nitrogen_factor):
    nh4_n_value = nh4_concentration / nitrogen_factor
    return nh4_n_value

def calculate_po4_p(po4_concentration, phosphorus_factor):
    po4_p_value = po4_concentration / phosphorus_factor
    return po4_p_value

def calculate_alkalinity(Vacid ,Nacid, Vsample):
    alkalinity_value = Vacid*Nacid*50*1000/Vsample
    return alkalinity_value

def absorption_analysis(initial_concentration, final_concentration, time):
    absorption_rate = (initial_concentration - final_concentration) / time
    return absorption_rate

def chloride_determination(sample_volume, silver_nitrate_volume, silver_nitrate_concentration):
    chloride_concentration = (silver_nitrate_volume * silver_nitrate_concentration * 35.45 * 1000) / sample_volume
    return chloride_concentration

def manganese_determination(initial_volume, final_volume, molarity_of_EDTA):
    manganese_concentration = ((initial_volume - final_volume) * molarity_of_EDTA * 54.93 * 1000) / sample_volume
    return manganese_concentration

def iron_determination(volume_of_sample, volume_of_KMnO4, molarity_of_KMnO4, factor):
    iron_concentration = ((volume_of_KMnO4 * molarity_of_KMnO4 * 5.6 * 10) / volume_of_sample) * factor
    return iron_concentration

def hardness_determination(volume_of_sample, EDTA_consumption, molarity_of_EDTA):
    hardness_determination = ((EDTA_consumption * molarity_of_EDTA) * 1000) / volume_of_sample
    return hardness_determination

def acidity_analysis(acid_volume, base_volume, normality_of_base, sample_volume):
    acidity_value = ((acid_volume - base_volume) * normality_of_base * 50 * 1000) / sample_volume
    return acidity_value

def toc_analysis(carbonate_concentration, carbon_factor):
    toc_value = carbonate_concentration / carbon_factor
    return toc_value

def dissolved_oxygen_analysis(oxygen_initial, oxygen_final, factor, volume):
    dissolved_oxygen_value = (oxygen_initial - oxygen_final) * factor * 8.36 * 1000 / volume
    return dissolved_oxygen_value

def calculate_oilandgres(last_gram, tare_of_balloon, volume):
    oilandgres_result = (last_gram - tare_of_balloon) * 1000 / volume
    return oilandgres_result

def ammonia_determination(sample_titration,witness_titration,sample_mL):
    ammonia_determination =((sample_titration - witness_titration)*280/sample_mL)
    return ammonia_determination

def fahrenheit_to_celsius(fahrenheit):
    """Convert temperature from Fahrenheit to Celsius."""
    return (fahrenheit - 32) * 5/9

def kelvin_to_celsius(kelvin):
    """Convert temperature from Kelvin to Celsius."""
    return kelvin - 273

def kg_per_cubic_meter_to_mg_per_liter(density_kg_per_m3):
    """Convert density from kg/m³ to mg/L."""
    return density_kg_per_m3 * 1e6

def plot_graph(x_values, y_values, x_label, y_label, title, chart_type='line'):
    if chart_type == 'line':
        plt.plot(x_values, y_values, marker='o')
    elif chart_type == 'bar':
        plt.bar(x_values, y_values)
    elif chart_type == 'scatter':
        plt.scatter(x_values, y_values)
    elif chart_type == 'pie':
        plt.pie(y_values, labels=x_values, autopct='%1.1f%%')


    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.show()

def main():
    print("Environmental Engineering Analysis Calculator")

    while True:
        print("\n1. COD Analysis")
        print("2. BOD Analysis")
        print("3. NH4-N Analysis")
        print("4. PO4-P Analysis")
        print("5. Alkalinity Analysis")
        print("6. Absorption Analysis")
        print("7. Chloride Determination")
        print("8. Manganese Determination")
        print("9. Iron Determination")
        print("10. Hardness Determination")
        print("11. Acidity Analysis")
        print("12. TOC Analysis")
        print("13. Dissolved Oxygen Analysis")
        print("14. oil-gres Analysis")
        print("15. Ammonia Determination")
        print("16. Unit Conversion For Temperature ")
        print("17. kg/m^3 to mg/l")
        print("18. Exit")


        user_choice = input("Select an option (1-18): ")

        if user_choice == '1':

            oxygen_initial = float(input("Enter initial dissolved oxygen concentration (mg/L): "))
            oxygen_final = float(input("Enter final dissolved oxygen concentration (mg/L): "))
            factor = float(input("Enter correction factor: "))
            volume = float(input("Enter solution volume (liters): "))
            cod_result = calculate_cod(oxygen_initial, oxygen_final, factor, volume)
            print(f"\nCOD Value: {cod_result} mg/L")


            chart_type = input("Select chart type (line/bar/scatter/pie): ")


            plot_graph([oxygen_initial, oxygen_final], [0, cod_result], 'Dissolved oxygen concentraiton(mg/L)', 'COD (mg/L)', 'COD Calculation Graph', chart_type)


        elif user_choice == '2':

            oxygen_initial = float(input("Enter initial dissolved oxygen concentration (mg/L): "))
            oxygen_final = float(input("Enter final dissolved oxygen concentration (mg/L): "))
            dilution_factor = float(input("Enter dilution factor: "))
            incubation_days = float(input("Enter incubation days: "))
            bod_result = calculate_bod(oxygen_initial, oxygen_final, dilution_factor, incubation_days)
            print(f"\nBOD Value: {bod_result} mg/L")


            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([oxygen_initial, oxygen_final], [0, bod_result], 'Dissolved oxygen concentration(mg/L)', 'BOD (mg/L)', 'BOD Calculation Graph', chart_type)


        elif user_choice == '3':

            nh4_concentration = float(input("Enter ammonium concentration (mg/L): "))
            nitrogen_factor = float(input("Enter nitrogen factor: "))
            nh4_n_result = calculate_nh4_n(nh4_concentration, nitrogen_factor)
            print(f"\nNH4-N Value: {nh4_n_result} mg/L")

            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([nh4_concentration, nitrogen_factor], [0, nh4_n_result], 'Nitrogen Value(mg/L)', 'NH4-N Result(mg/L)', 'NH4-N Calculation Graph', chart_type)



        elif user_choice == '4':

            po4_concentration = float(input("Enter phosphate concentration (mg/L): "))
            phosphorus_factor = float(input("Enter phosphorus factor: "))
            po4_p_result = calculate_po4_p(po4_concentration, phosphorus_factor)
            print(f"\nPO4-P Value: {po4_p_result} mg/L")

            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([po4_concentration, phosphorus_factor], [0, po4_p_result], 'po4 Concentration', 'Phosphorus Factor', 'po4-p Result', chart_type)




        elif user_choice == '5':

            Vacid= float(input("Enter V acid value (mg/L): "))
            Nacid= float(input("Enter N acid value : "))
            Vsample=float(input("Enter V sample value: "))
            alkalinity_result = calculate_alkalinity(Vacid,Nacid,Vsample)
            print(f"\nAlkalinity Value: {alkalinity_result} mg/L")
            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([Vsample, Vacid], [0, alkalinity_result], 'Volume Of Sample', 'Volume Of Acid', 'Alkalinity Result', chart_type)

        elif user_choice == '6':

            initial_concentration = float(input("Enter initial concentration: "))
            final_concentration = float(input("Enter final concentration: "))
            time = float(input("Enter time: "))
            absorption_rate, x_values, y_values = absorption_analysis(initial_concentration, final_concentration, time)
            print(f"\nAbsorption Rate: {absorption_rate} mg/L/min")

            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([initial_concentration, final_concentration], [0, absorption_rate], 'Concentration Value', 'Adsorption(mg/L)', 'Adsorption Calculation Graph', chart_type)

        elif user_choice == '7':

            sample_volume = float(input("Enter sample volume (L): "))
            silver_nitrate_volume = float(input("Enter silver nitrate volume (mL): "))
            silver_nitrate_concentration = float(input("Enter silver nitrate concentration (mol/L): "))
            chloride_result = chloride_determination(sample_volume, silver_nitrate_volume, silver_nitrate_concentration)
            print(f"\nChloride Concentration: {chloride_result} mg/L")

            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([silver_nitrate_volume, silver_nitrate_concentration], [0, chloride_result], 'Silver Nitrate Value(mg/L)', 'Chloride Value(mg/L)', 'Chloride Determination Calculation Graph', chart_type)


        elif user_choice == '8':

            initial_volume = float(input("Enter initial volume (mL): "))
            final_volume = float(input("Enter final volume (mL): "))
            molarity_of_EDTA = float(input("Enter molarity of EDTA (mol/L): "))
            manganese_result = manganese_determination(initial_volume, final_volume, molarity_of_EDTA)
            print(f"\nManganese Concentration: {manganese_result} mg/L")

            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([initial_volume, final_volume], [0, manganese_result], 'Volume(mg/L)', 'Manganese Value (mg/L)', 'Manganese Determination Calculation Graph', chart_type)

        elif user_choice == '9':

            volume_of_sample = float(input("Enter volume of sample (mL): "))
            volume_of_KMnO4 = float(input("Enter volume of KMnO4 (mL): "))
            molarity_of_KMnO4 = float(input("Enter molarity of KMnO4 (mol/L): "))
            factor = float(input("Enter factor: "))
            iron_result = iron_determination(volume_of_sample, volume_of_KMnO4, molarity_of_KMnO4, factor)
            print(f"\nIron Concentration: {iron_result} mg/L")

            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([volume_of_KMnO4, volume_of_sample], [0, iron_result], 'Volume(mg/L)', 'Iron Value(mg/L)', 'Iron  Determination Calculation Graph', chart_type)

        elif user_choice == '10':

            volume_of_sample = float(input("Enter volume of sample: "))
            EDTA_consumption = float(input("Enter EDTA consumption: "))
            molarity_of_EDTA = float(input("Enter molarity of EDTA: "))
            hardness_result = hardness_determination(volume_of_sample, EDTA_consumption, molarity_of_EDTA)
            print(f"\nHardness value: {hardness_result} mg/L")

            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([volume_of_sample, molarity_of_EDTA], [0, hardness_result], 'Volume(mg/L)', 'Hardness Determination(mg/L)', 'Hardness  Determinaiton Calculation Graph', chart_type)





        elif user_choice == '11':

            acid_volume = float(input("Enter volume of acid (mL): "))
            base_volume = float(input("Enter volume of base (mL): "))
            normality_of_base = float(input("Enter normality of base: "))
            sample_volume = float(input("Enter sample volume (mL): "))
            acidity_result = acidity_analysis(acid_volume, base_volume, normality_of_base, sample_volume)
            print(f"\nAcidity Value: {acidity_result} mg/L")
            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([acid_volume, base_volume], [0, sample_volume], 'Acidity Result (mg/L)', 'Sample Volume (mL)', 'Acidity Analysis Calculation Graph', chart_type)

        elif user_choice == '12':

            carbonate_concentration = float(input("Enter carbonate concentration (mg/L): "))
            carbon_factor = float(input("Enter carbon factor: "))
            toc_result = toc_analysis(carbonate_concentration, carbon_factor)
            print(f"\nTOC Value: {toc_result} mg/L")

            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([carbonate_concentration, carbon_factor], [0, toc_result], 'Carbonate concentration (mg/L)', 'TOC result (mg/l)', 'TOC Analysis Calculation Graph', chart_type)

        elif user_choice == '13':

            oxygen_initial = float(input("Enter initial dissolved oxygen concentration (mg/L): "))
            oxygen_final = float(input("Enter final dissolved oxygen concentration (mg/L): "))
            factor = float(input("Enter correction factor: "))
            volume = float(input("Enter solution volume (liters): "))
            dissolved_oxygen_result = dissolved_oxygen_analysis(oxygen_initial, oxygen_final, factor, volume)
            print(f"\nDissolved Oxygen Value: {dissolved_oxygen_result} mg/L")

            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([oxygen_initial, oxygen_final], [0, volume], 'Dissolved Oxygen Concentration (mg/L)', 'Solution Volume (Liters)', 'Dissolved Oxygen Analysis Calculation Graph', chart_type)

        elif user_choice == '14':

            last_gram = float(input("Last weight glass destialiton balloon: "))
            tare_of_balloon = float(input("Tare weight of glass destilation balloon: "))
            volume = float(input("Enter volume: "))
            oilandgres_result = calculate_oilandgres(last_gram, tare_of_balloon, volume)
            print(f"\ngres-oil value: {oilandgres_result} mg/L")

            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([last_gram, volume], [0, oilandgres_result], 'Last weight', 'volume', 'Oil-Gres Result', chart_type)


        elif user_choice == '15':
            sample_titration = float(input("Enter sample titraiton value(mg/L): "))
            witness_titration = float(input("Enter witness titration value (mg/L): "))
            sample_mL = float(input("Enter sample value(mL): "))
            ammonia_determination_result = ammonia_determination(sample_titration, witness_titration,sample_mL)
            print(f"\nAmmonia Determination Value: {ammonia_determination_result} mg/L")
            chart_type = input("Select chart type (line/bar/scatter/pie): ")
            plot_graph([sample_titration, witness_titration], [0, sample_mL], 'Titration Value(mg/L)', 'Sample Value (mg/L)', 'Ammonia Determination Calculation Graph', chart_type)


        elif user_choice == '16':
            unit = input("Enter the temperature unit (Fahrenheit or Kelvin): ")
            temperature = float(input("Enter the temperature value: "))


            if unit == 'Fahrenheit':
             celsius_temperature = fahrenheit_to_celsius(temperature)
             print(f"{temperature} Fahrenheit is {celsius_temperature:.2f} Celsius")
            elif unit == 'Kelvin':
             celsius_temperature = kelvin_to_celsius(temperature)
             print(f"{temperature} Kelvin is {celsius_temperature:.2f} Celsius")
            else:
             print("Invalid unit. Please enter 'F' or 'K' for Fahrenheit or Kelvin conversion.")

        elif user_choice == '17':
            density_kg_per_m3 = float(input("Enter the density in kg/m³: "))
            density_mg_per_liter = kg_per_cubic_meter_to_mg_per_liter(density_kg_per_m3)
            print(f"{density_kg_per_m3} kg/m³ is equal to {density_mg_per_liter:.2f} mg/L")

        elif user_choice == '18':
            print("Exiting the program.")
            break

        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
